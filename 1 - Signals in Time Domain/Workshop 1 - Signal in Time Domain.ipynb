```python
import numpy as np
import matplotlib.pyplot as plt

fs = 44100 # 44,100Hz sampling frequency
ts = 1/fs
t = np.arange(-1., 1., ts) # A linearlyspaced array with step ts
f = 2 # 2 Hz
fig_2, axes = plt.subplots(2,2, sharex='all', sharey='all', figsize=(18,18))

# Your code goes here
xt1 = np.cos(2 * np.pi * t)
xt2 = 0.5 * np.cos(2 * np.pi * t)
xt3 = -np.sin(2 * np.pi*t)
xt4 = np.cos(4 * np.pi * t)

axes[0, 0].plot(t,xt1)
axes[0, 0].set_xlabel('$t$')
axes[0, 0].set_ylabel('$x(t)$')
axes[0, 0].grid(True)
axes[0, 0].title.set_text('$\cos(2\pi t)$')

axes[0, 1].plot(t,xt2)
axes[0, 1].set_xlabel('$t$')
axes[0, 1].set_ylabel('$x(t)$')
axes[0, 1].grid(True)
axes[0, 1].title.set_text('$0.5 \cos(2\pi t)$')

axes[1, 0].plot(t,xt3)
axes[1, 0].set_xlabel('$t$')
axes[1, 0].set_ylabel('$x(t)$')
axes[1, 0].grid(True)
axes[1, 0].title.set_text('$-\sin(2\pi t)$')

axes[1, 1].plot(t,xt4)
axes[1, 1].set_xlabel('$t$')
axes[1, 1].set_ylabel('$x(t)$')
axes[1, 1].grid(True)
axes[1, 1].title.set_text('$\cos(4\pi t)$')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

fs = 4.4e4  # sampling frequency
ts = 1 / fs
t = np.arange(-1., 1., ts)
f = 5  # 5Hz

# Your code goes here
c_1 = 0.15
r_1 = 2
theta_1 = 0

c_2 = 0.15
r_2 = -2
theta_2 = np.pi / 2

w0 = 2 * np.pi * f

alpha = 1.2
xt_1 = c_1 * (np.exp(r_1 * t)) * np.cos(w0 * t + theta_1)
xt_2 = c_2 * (np.exp(r_2 * t)) * np.cos(w0 * t + theta_2)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(18, 18))

ax1.plot(t, xt_1, label='$0.15 e ^{2t} \cos(10\pi t)$')
ax2.plot(t, xt_2, label='$0.15 e ^{2t} \cos(10\pi t)$')

ax1.set_xlabel("t")
ax1.set_ylabel("x(t)")
ax1.title.set_text('$x(t) = 0.15 e^{2t} \cos(\omega t)$')
ax1.grid(True)

ax2.set_xlabel("t")
ax2.set_ylabel("x(t)")
ax2.title.set_text('$x(t) = 0.15 e^{-2t} \cos(\omega t + \pi / 2)$')
ax2.grid(True)

plt.show()
```

# Task 1.7

```python
import numpy as np
import matplotlib.pyplot as plt

def x(t):
    if (t < 0.):
        return 0.
    elif (t < 1.):
        return 1.
    elif (t < 2.):
        return 2. -t
    else:
        return 0.
fs = 4.4e4 # 44,000Hz sampling frequency
ts = 1/fs
t = np.arange(-3.5,3.5,ts) # A linearlyspaced array with step ts
fig, axes = plt.subplots(7,1, sharey='all', figsize=(10,15))
# x(t)
axes[0].plot(t, [x(t_) for t_ in t])
axes[0].set_xlabel('$t$')
axes[0].set_ylabel('$x(t)$')
axes[0].grid(True)
# Your code goes here

# x(t-1)
axes[1].plot(t, [x(t_) for t_ in t-1])
axes[1].set_xlabel('$t$')
axes[1].set_ylabel('$x(t-1)$')
axes[1].grid(True)

# x(t+1)
axes[2].plot(t, [x(t_) for t_ in t+1])
axes[2].set_xlabel('$t$')
axes[2].set_ylabel('$x(t+1)$')
axes[2].grid(True)

# x(-t)
axes[3].plot(t, [x(t_) for t_ in (-t)])
axes[3].set_xlabel('$t$')
axes[3].set_ylabel('$x(-t)$')
axes[3].grid(True)

# x(-t+1)
axes[4].plot(t, [x(t_) for t_ in (-t+1)])
axes[4].set_xlabel('$t$')
axes[4].set_ylabel('$x(-t+1)$')
axes[4].grid(True)

# x(3t/2)
axes[5].plot(t, [x(t_) for t_ in (3*t/2)])
axes[5].set_xlabel('$t$')
axes[5].set_ylabel('$x(3t/2)$')
axes[5].grid(True)

# x(3t/2 +1)
axes[6].plot(t, [x(t_) for t_ in (3*t/2 + 1)])
axes[6].set_xlabel('$t$')
axes[6].set_ylabel('$x(3t/2 + 1)$')
axes[6].grid(True)
plt.show()
```

# Task 1.8

```python
import numpy as np
import matplotlib.pyplot as plt


fs = 100# 100Hz sampling frequency
ts = 1/fs
t = np.arange(-1., 1., ts)# A linearlyspaced array with step ts
fig, axes = plt.subplots(2,1, figsize=(10,10))
f = 2 # 2 Hz
omega0 = 2*np.pi*f
xt = 0.75 + 2.*np.cos(omega0*t) + 1.*np.cos(3*omega0*t)
Xf = np.fft.fft(xt)
freq = np.fft.fftfreq(t.shape[-1], d=ts)
axes[0].plot(t,xt)
axes[0].set_xlabel('$t$ in s')
axes[0].set_ylabel('$x(t)$')
valsubrange = np.concatenate((np.arange(0,21,1), np.arange(-1,-21,-1)))
freqsubrage = np.concatenate((np.arange(0,21,1), np.arange(-1,-21,-1)))
axes[1].stem(freq[freqsubrage], Xf.real[valsubrange]/len(t))
axes[1].set_xlabel('$f$ in Hz')
axes[1].set_ylabel('$\mathfrak{Re}(X(j\omega))$')
plt.xticks(np.arange(-10,11))
plt.show()
```
